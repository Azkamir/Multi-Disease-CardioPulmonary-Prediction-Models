Imports

```python
import numpy as np
import pandas as pd
# import matplotlib.pyplot as plt
```

1-Reading Dataset from csv
2-Separating features coloumn and saving them in Variable X
3-separaring labels coloumn and saving it in Variable Y

```python
dataset=pd.read_csv("stroke dataset.csv")
from sklearn.utils import resample
dataset=resample(dataset,replace=True,n_samples=15000,random_state=1)
print(dataset.head)
X=dataset.iloc[:,1:-1].values
Y=dataset.iloc[:,-1].values
```

1-filling null values if there is any , with SimpleImputer class and update the Variable X
2-Using Label Encoder class to change abscent and  present into 0,1 for better computations and training

```python
from sklearn.preprocessing import LabelEncoder
EncodeClassLabels=LabelEncoder()
X1=EncodeClassLabels.fit_transform(X[:,[4]])
c=np.array(X1)
X[:,[4]]=c.reshape(len(c),1)

X2=EncodeClassLabels.fit_transform(X[:,[5]])
c2=np.array(X2)
X[:,[5]]=c2.reshape(len(c2),1)


X3=EncodeClassLabels.fit_transform(X[:,[6]])
c3=np.array(X3)
X[:,[6]]=c3.reshape(len(c3),1)


X4=EncodeClassLabels.fit_transform(X[:,[9]])
c4=np.array(X4)
X[:,[9]]=c4.reshape(len(c4),1)


X5=EncodeClassLabels.fit_transform(X[:,[0]])
c5=np.array(X5)
X[:,[0]]=c5.reshape(len(c5),1)

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
X = imputer.fit_transform(X)
print(X[:100])
```

splitting the data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)

```

Parameters value of model for cross validation

```python
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 80, num = 10)]
# # Number of features to consider at every split
# max_features = ['auto', 'sqrt']
# # Maximum number of levels in tree
# max_depth = [2,4]
# # Minimum number of samples required to split a node
# min_samples_split = [2, 5]
# # Minimum number of samples required at each leaf node
# min_samples_leaf = [1, 2]
# # Method of selecting samples for training each tree
# bootstrap = [True, False]
#
# params={'n_estimators': n_estimators,
#                'max_depth': max_depth,
#                'min_samples_split': min_samples_split,
#                'min_samples_leaf': min_samples_leaf,}
# print(params)

learn_rate = [ 0.1, 0.2, 0.3,1]
momentum = [0.0, 0.2, 0.4, 0.6, 0.8, 0.9]
param_grid=dict(learning_rate=learn_rate,n_estimators=n_estimators)
```

Transforming the values into range of values like -1 to 1 for better training

```python
# from sklearn.preprocessing import StandardScaler
# ss=StandardScaler()
# x_train=ss.fit_transform(x_train)
# x_test=ss.transform(x_test)
```

initializing randomforestclassifier
applying GridSearchCV(cross-validation)
and training using fit method

```python
from sklearn.ensemble import RandomForestClassifier as RFC
Mode=RFC(n_estimators=20,criterion="entropy")
from sklearn.ensemble import  AdaBoostClassifier
Model=AdaBoostClassifier(base_estimator=Mode)
from sklearn.model_selection import GridSearchCV
rfc=GridSearchCV(estimator=Model,param_grid =param_grid , cv = 3, verbose=2, n_jobs = 4)
rfc.fit(x_train,y_train)
```

making prediction on test data

```python
y_pred=rfc.predict(x_test)
```

checking accuracy using accuracy_score

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_pred)
print(ac)
```
